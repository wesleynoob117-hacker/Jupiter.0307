```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv(r'C:\Users\wesley_oliveira151\Downloads\dados_automacao_weg_bruto.csv')
print("--- Primeiras Linhas do Dataset Bruto ---")
display(dataset)
```

```python
dataset.columns = ['id', 'temperatura', 'pressao', 'status', 'genero', 'consumo_kwh']

dataset.head()
```

```python
# Agrupar e explorar a coluna de Status
grupado_status = dataset.groupby(['status']).size()
print("--- Agrupamento por Status ---")
print(grupado_status)
```

```python
# Agrupar e explorar a coluna de Status
grupado_genero = dataset.groupby(['genero']).size()
print("--- Agrupamento por Gênero ---")
display(grupado_genero)
```

```python
grupado_genero.plot.bar(color='black')
plt.title('Distribuição por Gênero do Operador')
plt.xlabel('Gênero')
plt.ylabel('Quantidade')
plt.show()
```
